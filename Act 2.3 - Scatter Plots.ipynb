***Scatter Plots***

-A circular dot is created in the plot to represent each point in the dataset.

-A relational plot that lets you discover trends. It describes the relationship between different variables.

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.shape
```

```python
diamonds.head()
```

```python
# To limit the number of obersvations, we are only interested in diamonds which have a premium or good cut 
# and follow a certian color pattern (D, F, and J).
# Furthermore, a random sample of 100 observarions was drawn.

diamonds = diamonds[
                diamonds.cut.isin(['Premium', 'Good']) &
                diamonds.color.isin(['D', 'F', 'J'])
                    ].sample(n=100, random_state=22)

diamonds.shape
```

```python
sns.set_style('dark')
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds) # looking at how carat is related to price
```

**Introducing more than two variables to a scatterplot**

Convey additional information from other variables through:

-Hue/Color

-Style

-Size

```python
sns.set_style('darkgrid')
```

```python
# Hue with categorical variable (cut)
sns.scatterplot(x='carat', y='price', hue='cut', data=diamonds, palette=['purple', '#55CCCC'], hue_order=['Premium', 'Good']) 
```

```python
# Hue with continuous variable (depth)
sns.scatterplot(x='carat', y='price', hue='depth', data=diamonds)
```

```python
# Changing style of dots
sns.scatterplot(x='carat', y='price', style='cut', data=diamonds, style_order=['Premium', 'Good'])
```

```python
# Changing size of dots
sns.scatterplot(x='carat', y='price', size='cut', data=diamonds, sizes=[150, 50], size_order=['Premium', 'Good'])
```

```python
#Combine hue, style, and size
sns.scatterplot(x='carat', y='price', hue='cut', style='color', data=diamonds, hue_order=['Premium', 'Good'], style_order=['F', 'D', 'J'])
```

**Additional helpful tools:**

-increase size of entire set of points: "s= "

-make points transparent if they overlap (1=solid, 0=invisible): "alpha= "

-change marker style: "marker= "

```python
sns.scatterplot(x='carat', y='price', style='cut', data=diamonds, style_order=['Premium', 'Good'], s=100, alpha=0.6)
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds, s=200, marker='*')
```
